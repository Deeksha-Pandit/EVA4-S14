```
import numpy as np
from PIL import Image
import random
import copy
import PIL
```

```
path = '/content/gdrive/My Drive/Mask_Rcnn/'
start = 388801 # -----------------------You should choose your numbers----------------------------------
black = np.zeros((224,224,3))

for i in range(87,101): #1 background image  -------------This is 1-100 == Same for all -------------

    bg = Image.open(f'{path}Background/bg{str(i)}.jpg')

    for j in  range(81,101): #1 foreground image  ---------------You should choose your numbers. Place only 20 images no flipped one-------
     
      fg = Image.open(f'{path}Foreground-Deeksha/fg{str(j)}.png')
      mask = Image.open(f'{path}Mask-Deeksha/mask{str(j)}.png')

      for k in range(1,21): #---------------------Randomly Placing 20 times(Same for all)--------------------------------
       
        r1 = random.randint(1, 80)
        r2 = random.randint(1, 80)
        bg1 = copy.deepcopy(bg)
        bg2 = copy.deepcopy(bg)
        fg1 = copy.deepcopy(fg)
        m1 = copy.deepcopy(mask)
        black_img1 = Image.fromarray(black,mode='RGB')
        black_img2 = Image.fromarray(black,mode='RGB')

        flipfg = fg1.transpose(PIL.Image.FLIP_LEFT_RIGHT) #flip image
        flipmask = m1.transpose(PIL.Image.FLIP_LEFT_RIGHT) #flip mask

        

        bg1.paste(fg1, (r1,r2),fg1)
        bg2.paste(flipfg,(r1,r2),flipfg)
        black_img1.paste(m1,(r1,r2), m1)
        black_img2.paste(flipmask,(r1,r2), flipmask)

        bg1.save(f"{path}Fg-Bg-Deeksha2/fg-bg{str(start)}.jpg")#------------Create Different folder-------------
        black_img1.save(f"{path}Fg-Bg-Mask-Deeksha2/fg-bg-mask{str(start)}.jpg") #-------------Create Different Path folder----

        start+=1

        bg2.save(f"{path}Fg-Bg-Deeksha2/fg-bg{str(start)}.jpg")#------------Create Different folder---------------
        black_img2.save(f"{path}Fg-Bg-Mask-Deeksha2/fg-bg-mask{str(start)}.jpg")#------------Create Different folder-----------

        start+=1
        print(start)
```
